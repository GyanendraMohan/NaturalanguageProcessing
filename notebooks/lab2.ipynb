```python
text = review_df[0]
print(text)
```

# 1. Perform stemming on the given text.

```python
from nltk.stem import PorterStemmer
from nltk.tokenize import word_tokenize
stemmer = PorterStemmer() # PoerterStemmer object
tokens = word_tokenize(text) # Tokenize the text
stemmed_tokens = [stemmer.stem(token) for token in tokens] # Stem each token
print(stemmed_tokens)
```

# 2. Perform lemmatization using WordNet.

```python
from nltk.stem import WordNetLemmatizer

lemmatizer = WordNetLemmatizer() # WordNetLemmatizer object
lemmatized_tokens = [lemmatizer.lemmatize(token) for token in tokens] # Lemmatize each token
print(lemmatized_tokens)   
```

# 3. Remove repeating characters from words.

```python
import re
import string

def remove_repeating_characters(word):
    return re.sub(r'(.)\1+', r'\1\1', word)

def remove_punctuation(word):
    return re.sub(f"[{re.escape(string.punctuation)}]", "", word)

def process_text(text):
    tokens = word_tokenize(text)
    tokens = [remove_punctuation(token) for token in tokens if token not in string.punctuation]
    stemmed_tokens = [stemmer.stem(token) for token in tokens]
    lemmatized_tokens = [lemmatizer.lemmatize(token) for token in tokens]
    processed_tokens = [remove_repeating_characters(token) for token in tokens]
    return processed_tokens

processed_text = process_text(text)
print(processed_text)
```

# 4. Replacing Words Using Regular Expressions
Definition:
Replace incorrect or unwanted words using pattern matching.

```python
def replace_word(text, target_word, replacement_word):
    pattern = re.compile(re.escape(target_word), re.IGNORECASE)
    return pattern.sub(replacement_word, text)
replaced_text = replace_word(text, "reviewers", "critics")
print(replaced_text)
```

# 5. Spelling Correction
Definition:
Automatically correct spelling errors in text.

```python
from textblob import TextBlob
def correct_spelling(replace_text):
    blob = TextBlob(replace_text)
    print(blob)
    corrected_text = blob.correct()
    return str(corrected_text)
corrected_text = correct_spelling(replaced_text)
print("corrected text:",corrected_text)
```

# 6. Replacing Synonyms
Definition:
Replace a word with another word having similar meaning.

```python
from nltk.corpus import wordnet
def replace_synonyms(text, target_word):
    synonyms = wordnet.synsets(target_word)
    if synonyms:
        synonym = synonyms[0].lemmas()[0].name()
        print(f"Replacing '{target_word}' with its synonym '{synonym}'")
        pattern = re.compile(re.escape(target_word), re.IGNORECASE)
        return pattern.sub(synonym, text)
    return text
synonym_replaced_text = replace_synonyms(corrected_text, "watching")
print("synonym replaced text:", synonym_replaced_text)
```

# 8. Translation
Definition:
Translate text from one language to another.

```python
from googletrans import Translator
import asyncio
```

```python
async def translate_text(text, dest_language):
    translator = Translator()
    translation = await translator.translate(text, dest=dest_language)
    return translation.text

translated_text = await translate_text(synonym_replaced_text, 'fr')
print("translated text:", translated_text)
```
